```python
from square import square # import the square function we wrote

# use the range function to generate the integers 1 through 10
xs = list(range(1, 11))
# square each element in xs
xs_squared = [square(x) for x in xs]
print("List of x values:", xs)
print("List of x^2 values:", xs_squared)
```

```python
import matplotlib.pyplot as plt

plt.scatter(xs, xs_squared, color='black')  # black dots
plt.title("scatter plot: y = x²")
plt.xlabel("x")
plt.ylabel("x²")
plt.show()
# (no need to match the axis labels/plot title exactly)
```
